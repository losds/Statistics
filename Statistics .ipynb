Agafa un conjunt de dades de tema esportiu que t'agradi i selecciona un atribut del conjunt de dades. Calcula la moda, la mediana, la desviació estàndard i la mitjana aritmètica. 

```python
import pandas as pd
```

```python
dfn=pd.read_csv('mlbBat10.txt',sep='\t')
dfn
```

```python
moda=dfn['G'].mode()
```

```python
moda
```

```python
mediana=dfn['G'].median()
mediana
```

```python
desviacion_estandart=dfn['G'].std()
desviacion_estandart
```

```python
mediana_aritmetica=dfn['G'].sum()/dfn['G'].shape[0]
mediana_aritmetica
```
